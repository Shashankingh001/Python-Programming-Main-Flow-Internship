```python
import tkinter as tk
from tkinter import messagebox

class GroceryStore:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Grocery Store Billing System",)
        self.root.config(background="white")
        self.root.geometry("400x300")
        self.inventory = {"Apple": 100.00, "Banana": 50.00, "Orange": 60.00, "Mango": 20.00}
        self.bill_items = {}

        # Create labels and entries
        label1 = tk.Label(self.root, borderwidth=3, bg='white', highlightbackground="white", highlightcolor="white", highlightthickness=1, text="Item Name:",font=('arial',9,'bold'))
        label1.pack()
        self.entry1 = tk.Entry(self.root,borderwidth=3, bg='#a3f5ee', highlightbackground="white", highlightcolor="white", highlightthickness=1)
        self.entry1.pack()
        label2 = tk.Label(self.root,borderwidth=5, bg='white', highlightbackground="white", highlightcolor="white", highlightthickness=1, text="Quantity:",font=('arial',9,'bold'))
        label2.pack()
        self.entry2 = tk.Entry(self.root,borderwidth=3, bg='#a3f5ee', highlightbackground="white", highlightcolor="white", highlightthickness=1)
        self.entry2.pack()

        # Create buttons
        button1 = tk.Button(self.root,width=8,height=1, text="Add to Bill", borderwidth=3, bg='#54bc86', highlightbackground="white", highlightcolor="white", highlightthickness=1, command=self.add_to_bill)
        button1.pack()
        button2 = tk.Button(self.root,width=8,height=1, text="Finish Bill", borderwidth=3, bg='#54bc86', highlightbackground="white", highlightcolor="white", highlightthickness=1,command=lambda: self.finish_bill(self.root))
        button2.pack()

        # Create text box
        self.text_box = tk.Text(self.root,borderwidth=3, height=10, width=40,bg='white',font=('arial',8,'bold'))
        self.text_box.pack()

    def add_to_bill(self):
        item_name = self.entry1.get()
        quantity = int(self.entry2.get())
        if item_name in self.inventory:
            price = self.inventory[item_name]
            total_price = price * quantity
            self.bill_items[item_name] = {"quantity": quantity, "price": price, "total_price": total_price}
            self.text_box.delete(1.0, tk.END)
            for item, details in self.bill_items.items():
                self.text_box.insert(tk.END, f"{item}: {details['quantity']} x Rs{details['price']:.2f} = Rs{details['total_price']:.2f}\n")
        else:
            messagebox.showerror("Item Not Found", f"Item '{item_name}' not found in the inventory.")

    def finish_bill(self, bill_window):
        total_amount = sum(item["total_price"] for item in self.bill_items.values())
        payment_window = tk.Toplevel(bill_window)
        payment_window.title("Make Payment")
        payment_window.config(background="white")
        payment_window.geometry("300x200")

        # Create labels and entries
        label1 = tk.Label(payment_window, text="Total Amount:",bg='white', highlightbackground="white", highlightcolor="white", highlightthickness=1,font=('arial',9,'bold'))
        label1.pack()
        label2 = tk.Label(payment_window, text=f"Rs{total_amount:.2f}",bg='#a3f5ee', highlightbackground="white", highlightcolor="white", highlightthickness=1,font=('arial',8,'bold'))
        label2.pack()
        label3 = tk.Label(payment_window, text="Payment Amount:",bg='white', highlightbackground="white", highlightcolor="white", highlightthickness=1,font=('arial',9,'bold'))
        label3.pack()
        entry1 = tk.Entry(payment_window, bg='#a3f5ee', highlightbackground="white", highlightcolor="white", highlightthickness=1)
        entry1.pack()

        # Create buttons
        button1 = tk.Button(payment_window, text="Make Payment", borderwidth=2.5, font=('arial',9,'bold'), bg= "#54bc86", fg="black", padx=15, command=lambda: self.make_payment(entry1, payment_window, total_amount))
        button1.pack()

    def make_payment(self, entry1, payment_window, total_amount):
        payment_amount = float(entry1.get())
        if payment_amount >= total_amount:
            change = payment_amount - total_amount
            messagebox.showinfo("Payment Successful", f"Payment of Rs{payment_amount:.2f} successful. Change: Rs{change:.2f}")
            payment_window.destroy()
        else:
            messagebox.showerror("Insufficient Payment", f"Payment of Rs{payment_amount:.2f} is insufficient. Please pay Rs{total_amount:.2f} or more.")

    def run(self):
        self.root.mainloop()

if __name__ == "__main__":
    store = GroceryStore()
    store.run()
```
